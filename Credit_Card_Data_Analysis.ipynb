Importing libraries and reading the data file.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
%matplotlib inline

# Read the data
data = pd.read_csv('creditcard.csv')
```

Exploring the data file.

```python
#Checking if there is any missing data.
data.info()
```

```python
# Checking the data
data.describe()
```

```python
data.head()
data.tail()
```

```python
#Mappings
fraud_mapping = {0: 'Not Fraud', 1: 'Fraud'}
data["Class"] = data["Class"].map(fraud_mapping)
```

```python
# Plotting the pie chart to see the percentage of fraud and non-fraud cases
counts = data['Class'].value_counts()
plt.pie(counts, labels=counts.index, autopct='%1.2f%%', shadow=True, startangle=90)
plt.title('Credit Card Fraud Class Percentage')
plt.show()
```

```python
# Checking how much transaction amounts differ between fraud and non-fraud transactions
bins = 30
ax1 = sns.displot(data.Amount[data.Class == "Fraud"], bins = bins)
ax2 = sns.displot(data.Amount[data.Class == "Not Fraud"], bins = bins)
ax1.set(xlabel='Amount ($)', ylabel='Number of Transactions', yscale='log', title='Fraud')
ax2.set(xlabel='Amount ($)', ylabel='Number of Transactions', yscale='log', title='Not Fraud')

plt.xlabel("Amount ($)")
plt.ylabel("Number of Transactions")
plt.yscale('log')
plt.show()
```

```python
# Checking how fraud and non fraud distributions differ from each other on symlog scale.
v_features = data.iloc[:,1:29].columns
plt.figure(figsize=(12,28*4))
gs = gridspec.GridSpec(28, 1)
for i, cn in enumerate(data[v_features]):
    ax = plt.subplot(gs[i])
    sns.histplot(data=data[data['Class'] == "Fraud"], x=cn, bins=50, kde=True, color='red', label='Fraud')
    sns.histplot(data=data[data['Class'] == "Not Fraud"], x=cn, bins=50, kde=True, color='blue', label='Not Fraud')
    plt.xlabel('')
    plt.yscale('symlog')
    plt.title('Histogram of feature: ' + str(cn))
plt.show()
```
